```python
import numpy as np
import re
from scipy.spatial import distance

f = open('sentences.txt')
a = re.split('[^a-z]', f.read().lower())
f.close()

f = open('sentences.txt')
n = len(f.read().split('\n'))
f.close()

a = list(set(a))
a.remove('')

d = {}
for i in range(len(a)):
    d[a[i]] = i

arr = np.zeros((n,len(d)))

i = 0
j = 0

f = open('sentences.txt')
while i < n:
    l = re.split('[^a-z]', f.readline().lower())
    for sent_w in l:
        for word in d:
            if word == sent_w:
                arr[i][j] += 1
            j += 1
        j = 0
    i += 1

```
