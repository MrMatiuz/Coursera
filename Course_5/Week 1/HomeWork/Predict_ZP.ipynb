Теперь это больше похоже на белый шум; можно сказать, что есть стационарность

Теперь подберем параметры p, P, q, Q:

```python
plt.figure(figsize(15,20))

ax = plt.subplot(2,1,1)
sm.graphics.tsa.plot_acf(data.box_diff3[14:].values.squeeze(), lags=48, ax=ax)
plt.xticks(range(49))
plt.grid()
plt.show()

ax = plt.subplot(2,1,2)
sm.graphics.tsa.plot_pacf(data.box_diff3[14:].values.squeeze(), lags=48, ax=ax)
plt.xticks(range(49))
plt.grid()
plt.show()
```

* Q * S = 12   =>  Q = 1
* Т.к. Q > 0 => берем q < S  =>  q = 11
* P * S = 0  =>  P = 0
* Можно взять p = 47 или 39, но это слишком далекие значения и вероятнее всего связанные со случайностью, поэтому пока попробуем взять p = 11. Если при таком p модель будет не очень, тогда попробуем и большие, иначе при больном значении p будет много комбинаций, что сильно растянет подсчет.


```python
d = 2
D = 1
ps = range(12)
qs = range(12)
Ps = range(1)
Qs = range(2)
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(data.WAG_C_M_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12),
                                        enforce_stationarity=False,
                                        enforce_invertibility=False,
                                        trend='c').fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

```python
plt.figure(figsize(15,20))
plt.subplot(211)
best_model.resid[14:].plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[14:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[14:], 0)[1])
print("Критерий Дики-Фуллера: p=%e" % sm.tsa.stattools.adfuller(best_model.resid[14:])[1])
```

Остатки несмещены (подтверждается критерием Стьюдента) стационарны (подтверждается критерием Дики-Фуллера и визуально), автокоррелированы, что плохо (подтверждается критерием Льюнга-Бокса и коррелограммой).
Посмотрим, насколько хорошо модель описывает данные:
