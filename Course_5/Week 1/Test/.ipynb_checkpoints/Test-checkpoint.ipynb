# Test №1

```python
import pandas as pd
import statsmodels.api as sm
%pylab inline
```

```python
milk = pd.read_csv('monthly-milk-production.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)
```

```python
milk
```

```python
pylab.plot(milk)
```

```python
_, p, _, _, _, _= sm.tsa.stattools.adfuller(milk)
```

```python
round(p,2)
```

```python
milk['days'] =  [pd.Period(milk.index.astype(str)[i]).days_in_month for i in range(168)]
milk['milk_per_day'] = milk.milk / milk.days
```

```python
milk
```

```python
pylab.plot(milk.milk_per_day)
```

```python
round(milk.milk_per_day.sum(),2)
```

# Test №2

```python
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
```

```python
milk['daily_diff1'] = milk.milk_per_day.diff(1)
milk['daily_diff2'] = milk.milk_per_day.diff(12)
milk['daily_diff12'] = milk.daily_diff1.diff(12)
```

```python
milk.head(15)
```

```python
pylab.figure(figsize = (17,8))
pylab.plot(milk.daily_diff1, color='r')
pylab.plot(milk.daily_diff2, color='b')
pylab.plot(milk.daily_diff12, color='g')
pylab.grid()
pylab.legend(['несезонное диф.','сезонное диф.', 'несезонное диф. + сезонное диф.'])
pylab.show()
```

```python
print("p_val for diff1 =", adfuller(milk.daily_diff1[1:])[1])
print("p_val for diff2 =", adfuller(milk.daily_diff2[12:])[1])
print("p_val for diff12 =", adfuller(milk.daily_diff12[13:])[1])
```

```python
plt.figure(figsize(15,8))
sm.graphics.tsa.plot_acf(milk.daily_diff12[13:].values.squeeze(), lags=50)
plt.grid()
plt.show()
```

```python
plt.figure(figsize(15,8))
sm.graphics.tsa.plot_pacf(milk.daily_diff12[13:].values.squeeze(), lags=50)
plt.grid()
plt.show()
```
